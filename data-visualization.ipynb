```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import *
import numpy as np
from pylab import *
```

```python
sin = np.sin(np.linspace(-4,4,100))
fig, ax = plt.subplots(2, 4, figsize=(20,10))
ax[0,0].plot(sin)
ax[0,1].plot(range(100), sin)
ax[0,2].plot(np.linspace(-4,4,100), sin)
ax[0,3].plot(sin[::10], 'o', c='r')
ax[1,0].plot(sin, c ='r', lw=8)
ax[1,1].plot(sin, '--')
ax[1,2].plot(sin, lw=10)
ax[1,3].plot(sin[::10], '--o')
plt.tight_layout()
```
